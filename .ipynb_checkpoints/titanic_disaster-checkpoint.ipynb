## Classes Analysis

Bilet daƒüƒ±lƒ±mƒ±na bir g√∂z atalƒ±m.

‚ùì Bir `groupby()`,kullanarak, her sƒ±nƒ±f (1, 2 veya 3) i√ßin satƒ±lan bilet sayƒ±sƒ±nƒ± hesaplayan bir`pclass_df` dataframe olu≈ütur.

```python
pclass_df = (
    titanic_df
    .groupby("Pclass")
    .size()
    .reset_index(name="PassengerCount")
)

pclass_df
```

Sayƒ±ya bakmak √ßok g√∂rsel deƒüil, veriyi bir grafikle anlamlandƒ±rmayƒ± deneyelim.

‚ùì Bir √∂nceki soruda olu≈üturduƒüun `pclass_df` dataframe‚Äôini bir bar chart olarak plot et.

```python
pclass_df.plot(
    x="Pclass",
    y="PassengerCount",
    kind="bar",
    legend=False,
    title="Passenger Count by Ticket Class"
)

plt.xlabel("Ticket Class")
plt.ylabel("Number of Passengers")
plt.show()
```

≈ûimdi hayatta kalanlara **survivors** bakalƒ±m.

‚ùì Her yolcu sƒ±nƒ±fƒ±nƒ±n hayatta kalma oranƒ±nƒ±  *survival rate* g√∂steren bir bar chart plot et.
`0` ,  o sƒ±nƒ±fta kimsenin hayatta kalmadƒ±ƒüƒ±, `1` ise herkesin hayatta kaldƒ±ƒüƒ± anlamƒ±na gelir. 

```python
survival_by_class = (
    titanic_df
    .groupby("Pclass")["Survived"]
    .mean()
    .reset_index(name="SurvivalRate")
)

survival_by_class
```

```python
survival_by_class.plot(
    x="Pclass",
    y="SurvivalRate",
    kind="bar",
    legend=False,
    title="Survival Rate by Ticket Class"
)

plt.xlabel("Ticket Class")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

### Gender Analysis

`Sex`  s√ºtununa bir g√∂z atalƒ±m.

‚ùì Bu s√ºtunda kullanƒ±lan farklƒ± deƒüerleri kontrol etmek i√ßin [`pandas.Series.unique`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.unique.html) fonksiyonunu kullan.

```python
titanic_df["Sex"].unique()
```

‚ùì Her cinsiyetin hayatta kalma oranƒ±nƒ± *survival rate*  g√∂steren bir bar chart plot et.
Verilere g√∂re, hangi cinsiyetin hayatta kalma ihtimali daha y√ºksekti?

```python
survival_by_gender = (
    titanic_df
    .groupby("Sex")["Survived"]
    .mean()
    .reset_index(name="SurvivalRate")
)

survival_by_gender
```

Her cinsiyet i√ßin toplam yolcu sayƒ±sƒ±nƒ± ve toplam hayatta kalan sayƒ±sƒ±nƒ± birlikte g√∂sterdiƒüimiz daha ≈üƒ±k bir histogram olu≈üturalƒ±m.

‚ùì`Total` ve `Survived`  olmak √ºzere iki s√ºtunlu, `male` and `female` ≈üeklinde iki satƒ±rlƒ± bir `survivors_df` DataFrame olu≈ütur ve plot et.

```python
survival_by_gender.plot(
    x="Sex",
    y="SurvivalRate",
    kind="bar",
    legend=False,
    title="Survival Rate by Gender"
)

plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

```python
survivors_df = pd.DataFrame({
    "Total": titanic_df["Sex"].value_counts(),
    "Survived": titanic_df[titanic_df["Survived"] == 1]["Sex"].value_counts()
})

survivors_df
```

```python
survivors_df.plot(
    kind="bar",
    title="Total vs Survived Passengers by Gender"
)

plt.ylabel("Number of Passengers")
plt.show()
```

## Children

√ñnceki analiz ya≈ülarƒ± dikkate almƒ±yordu. Bir √ßocuƒüu bir yeti≈ükinden ayƒ±rmak ve bunun hayatta kalma oranlarƒ±nƒ± *survival rates* nasƒ±l etkilediƒüini g√∂rmek istiyoruz.

‚ùì Boolean indexing kullanarak yalnƒ±zca √ßocuk yolcularƒ±n satƒ±rlarƒ±nƒ± i√ßeren bir `children_df` olu≈ütur.

```python
children_df = titanic_df[titanic_df["Age"] < 18]
children_df.head()
```

‚ùì Titanic‚Äôte ka√ß √ßocuk vardƒ±?

```python
len(children_df)
```

‚ùìKa√ß √ßocuk hayatta kaldƒ±?

```python
children_df["Survived"].sum()
```

‚ùì Her kategori i√ßin( `male`, `female`, `children`).hayatta kalanlar / toplam oranƒ±nƒ± g√∂steren bir bar chart plot et. Unutma: `male` istatistiklerinden erkek √ßocuklarƒ±, `female` istatistiklerinden ise kƒ±z √ßocuklarƒ± √ßƒ±karman **substract** gerekiyor.

```python
total_male = len(titanic_df[titanic_df["Sex"] == "male"])
total_female = len(titanic_df[titanic_df["Sex"] == "female"])
total_children = len(children_df)

survived_male = titanic_df[(titanic_df["Sex"] == "male") & (titanic_df["Survived"] == 1)].shape[0]
survived_female = titanic_df[(titanic_df["Sex"] == "female") & (titanic_df["Survived"] == 1)].shape[0]
survived_children = children_df["Survived"].sum()

rates_df = pd.DataFrame({
    "SurvivalRate": [
        survived_male / total_male,
        survived_female / total_female,
        survived_children / total_children
    ]
}, index=["Male", "Female", "Children"])

rates_df
```

```python
rates_df.plot(
    kind="bar",
    legend=False,
    title="Survival Rate: Male vs Female vs Children"
)

plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

## [Opsiyonal] Big families

‚ùì Daha b√ºy√ºk ailelerin hayatta kalmasƒ± daha mƒ± zordu?

Burada, yolcunun aile b√ºy√ºkl√ºƒü√ºn√º g√∂steren yeni bir s√ºtunu `DataFrame` ine ekleyeceksin.
  

```python
# SENƒ∞N KODUN BURAYA

titanic_df["FamilySize"] = titanic_df["SibSp"] + titanic_df["Parch"] + 1
```

```python
titanic_df[["SibSp", "Parch", "FamilySize"]].head()
```

```python
family_survival = (
    titanic_df
    .groupby("FamilySize")["Survived"]
    .mean()
    .reset_index(name="SurvivalRate")
)

family_survival
```

```python
family_survival.plot(
    x="FamilySize",
    y="SurvivalRate",
    kind="bar",
    legend=False,
    title="Survival Rate by Family Size"
)

plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

```python
family_stats = (
    titanic_df
    .groupby("FamilySize")["Survived"]
    .agg(Total="count", SurvivalRate="mean")
    .reset_index()
    .sort_values("FamilySize")
)

family_stats
```

```python
ax = family_stats.plot(x="FamilySize", y="SurvivalRate", kind="bar", legend=False, title="Survival Rate by Family Size")
ax.set_xlabel("Family Size")
ax.set_ylabel("Survival Rate")
ax.set_ylim(0, 1)

# bar √ºst√ºne Total yaz
for i, row in family_stats.iterrows():
    ax.text(i, row["SurvivalRate"] + 0.02, f'n={int(row["Total"])}', ha='center', fontsize=9)

plt.show()
```

## [Opsiyonal] Distinguished titles

‚ùì Tahliye sƒ±rasƒ±nda √ºnvanƒ± (Mr, Mrs, Dr vb.) olan yolculara daha fazla √∂ncelik verildi mi?

Bunu incelemek i√ßin, bazƒ± string manip√ºlasyonlarƒ± yaparak her yolcu i√ßin √ºnvanƒ±nƒ± i√ßeren yeni bir s√ºtun olu≈ütur.

```python
# SENƒ∞N KODUN BURAYA

titanic_df["Title"] = titanic_df["Name"].str.extract(r",\s*([^\.]+)\.")
```

```python
titanic_df[["Name", "Title"]].head()
```

```python
titanic_df["Title"].value_counts()
```

```python
title_survival = (
    titanic_df
    .groupby("Title")["Survived"]
    .mean()
    .sort_values(ascending=False)
)

title_survival
```

```python
title_survival.plot(
    kind="bar",
    title="Survival Rate by Title"
)

plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

```python
title_stats = (
    titanic_df.groupby("Title")["Survived"]
    .agg(Total="count", SurvivalRate="mean")
    .reset_index()
    .sort_values(["SurvivalRate", "Total"], ascending=[False, False])
)

title_stats
```

```python
titanic_df["TitleGrouped"] = titanic_df["Title"].where(titanic_df["Title"].isin(["Mr", "Mrs", "Miss", "Master"]), "Rare")

title_grouped_stats = (
    titanic_df.groupby("TitleGrouped")["Survived"]
    .agg(Total="count", SurvivalRate="mean")
    .reset_index()
    .sort_values("SurvivalRate", ascending=False)
)

title_grouped_stats
```

```python
ax = title_grouped_stats.plot(x="TitleGrouped", y="SurvivalRate", kind="bar", legend=False, title="Survival Rate by Title (Grouped)")
ax.set_xlabel("Title Group")
ax.set_ylabel("Survival Rate")
ax.set_ylim(0, 1)

for i, row in title_grouped_stats.iterrows():
    ax.text(i, row["SurvivalRate"] + 0.02, f'n={int(row["Total"])}', ha='center', fontsize=10)

plt.show()
```
